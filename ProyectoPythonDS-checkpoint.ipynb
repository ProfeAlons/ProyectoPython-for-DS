```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar los datos de Tesla utilizando yfinance
tesla = yf.Ticker('TSLA')

# Obtener el historial de precios de las acciones
tesla_data = tesla.history(period="1y")  # Datos de 1 año

# Crear un gráfico de las acciones de Tesla
def make_graph(data):
    # Graficar el precio de cierre ajustado
    plt.figure(figsize=(10,6))
    plt.plot(data.index, data['Close'], label="Precio de Cierre", color='b')
    
    # Añadir título y etiquetas
    plt.title('Precio de las Acciones de Tesla (Último Año)', fontsize=14)
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Precio (USD)', fontsize=12)
    
    # Mostrar la leyenda y el gráfico
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función make_graph para crear el gráfico
make_graph(tesla_data)

```
